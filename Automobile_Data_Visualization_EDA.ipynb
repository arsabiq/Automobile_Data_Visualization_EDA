```python
import numpy as np
import pandas as pd
# basic visualization package - bar, line, pie, scatter, histogram
import matplotlib.pyplot as plt
# discriptive plots - kde, box, heatmap
import seaborn as sns
```

```python
auto_price = pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv")
```

```python
auto_price.head()
```

```python
auto_price.drop(['symboling','normalized-losses'], axis = 1, inplace = True)
```

```python
# replace '?' by np.nan
auto_df = auto_price.replace('?', np.nan)
```

```python
auto_df.isnull().sum()
```

```python
# drop the rows with missing data 
auto_df.dropna(inplace = True)
```

```python
auto_df.info()
```

```python
cols = ['bore','stroke','horsepower','peak-rpm','price']
auto_df[cols] = auto_df[cols].apply(pd.to_numeric)
```

##### Line Plot
    * Univariate or Bivariate (Relationship Plot)
    * Trend Analysis (Time Series) - Univariate Analysis

```python
auto_df['price'].sort_values().unique()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize = (10,6))
plt.plot(auto_df['price'].sort_values().unique())
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize = (10,6))
plt.plot(auto_df.sort_values(by = 'price')['price'], auto_df.sort_values(by = 'price')['city-mpg'],color = 'darkviolet')
plt.title('Price vs City Mileage')
plt.xlabel('Price')
plt.ylabel('City Mileage')
plt.show()
```

    * Interpretation: With Increase in vehicle price there is decrease in mileage.

#### Bar Plot
    * Bivariate / Univariate Analysis
    * Bar chart used by Categorical, Nominal, Categorical vs Value Counts/Numerical/Continious
    * Bar Plots are used to display counts (.value_counts()) of unique value for categorical dtype, Height of Bar represents count for each category.

```python
auto_df[['price']].loc[10:75,].plot.bar(figsize = (12,6),color = 'deepskyblue')
```

```python
auto_df['make'].value_counts()
```

```python
auto_df['make'].value_counts().plot.bar(figsize = (10,6))
```

##### Histogram

    * Continious Samples - Study the spread / distribution of Data
    * Univariate Analysis.

```python
plt.figure(figsize = (9,5))
# bins = num of intervals
# Price of vehicle is positively skewed
plt.hist(auto_df['price'], color = 'orange', bins = 10)
plt.title('Histogram Plot - Auto Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#### Distribution Plot

    * Histogram + KDE
    * Probability Density Function (PDF) -> Continious

```python
import warnings
warnings.filterwarnings('ignore')

fig = plt.figure(figsize = (9,6))
sns.distplot(auto_df['engine-size'], bins = 10, color = 'blue')
plt.show()
```

```python
fig = plt.figure(figsize = (9,6))
# hist = False => KDE Plot
sns.distplot(auto_df['engine-size'], bins = 10, color = 'blue', hist = False)
plt.show()
```

#### Scatter Plot

    * Scatter Plots is plotted between numerical values.
    * It shows relationship b/w x & y plots
    * Bivariate Analysis.  

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize = (9,6))
# marker = shape of scatter plot 's','*','d','^','>','<','o','.'
# alpha = color transperancy
sns.scatterplot(x = auto_df['city-mpg'], y = auto_df['price'], color = 'deepskyblue', marker = 's', alpha = 0.6)
plt.show()
```

#### Pie-Chart

```python
auto_df['make'].value_counts()
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (6,6))
# autopct = compute percentage %
plt.pie(auto_df['make'].value_counts(), labels = auto_df['make'].value_counts().index, autopct = '%0.0f%%')
plt.show()
```

###### BoxPlot
    * Distribution of Continious Data
    * It shows statistical distribution like - Quartiles (Q1, Q2, Q3) and Median (50% or Q2).
    * This plot is useful in showing Outliers.

```python
plt.figure(figsize = (7,3))
# orient = 'h' , 'h' -> Horizontal Orientation
sns.boxplot(auto_df['price'], orient = 'h')
plt.show()
```
